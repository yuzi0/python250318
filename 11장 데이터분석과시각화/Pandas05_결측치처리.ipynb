```python
#결측치처리
import numpy as np
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

```python
# 컬럼 (column) 설명
# survivied: 생존여부 (1: 생존, 0: 사망)
# pclass: 좌석 등급 (1등급, 2등급, 3등급)
# sex: 성별
# age: 나이
# sibsp: 형제 + 배우자 수
# parch: 부모 + 자녀 수
# fare: 좌석 요금
# embarked: 탑승 항구 (S, C, Q)
# class: pclass와 동일
# who: 남자(man), 여자(woman), 아이(child)
# adult_male: 성인 남자 여부
# deck: 데크 번호 (알파벳 + 숫자 혼용)
# embark_town: 탑승 항구 이름
# alive: 생존여부 (yes, no)
# alone: 혼자 탑승 여부
df.info()
```

```python
#널값이 있는 컬럼의 숫자를 출력 
df.isnull().sum()
```

```python
#isnull(), isna()는 결과가 동일하다. 편한 것을 사용한다. 
df.isna().sum()
```

```python
#각컬럼의 널값 숫자를 전부 더한다.
df.isnull().sum().sum()
```

```python
#널값이 아닌 컬럼의 숫자
df.notnull().sum()
```

```python
#나이가 널값인 경우만 필터링한다.
df.loc[df['age'].isnull()]
```

```python
#복사본을 만든다.
df1 = df.copy()
```

```python
df1.tail()
```

```python
#널값을 700으로 변경한다.
df1['age'].fillna(700).tail()
```

```python
#업데이트한 값으로 덮어쓰기를 한다. 
df1['age'] = df1['age'].fillna(700)
```

```python
df1 
```

```python
#다시 복사한다. 
df1 = df.copy()
df1.tail()
```

```python
#평균값으로 업데이트 한다.
df1['age'].fillna(df1['age'].mean()).tail()
```

```python
#중위값으로 업데이트 한다.
df1['age'].fillna(df1['age'].median()).tail()
```

```python
#널값이 있으면 삭제한다.
df1.dropna()
```

```python
#옵션을 지정할 수 있다. 
#전체 컬럼이 널값인 경우만 삭제한다.
df1.dropna(how='all')
```
